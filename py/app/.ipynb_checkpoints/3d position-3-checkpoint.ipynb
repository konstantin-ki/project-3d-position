```python
import matplotlib.pyplot as plt
import matplotlib.animation as an
from datetime import datetime
import random
# Graphics in retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
%matplotlib widget
```

```python
fig = plt.figure()
ax  =fig.add_subplot(1,1,1)
#plt.xticks(rotation ='vertical')

fig.suptitle('ГРАФИК ПОКАЗАНИЙ АКСЕЛЕРОМЕТРА')
ax.set_xlabel('ВРЕМЯ')
ax.set_ylabel('ЗНАЧЕНИЯ АКСЕЛЕРОМЕТРА')
 
x=[]
y=[]
xcaption=[]
cur_number=0;
 
def UpdateGraph(frame):
    #cur_datetime=datetime.now()
    global cur_number
    cur_number=cur_number+1
    x.append(cur_number)
    y.append(random.randint(5,10))
    #xcaption.append(cur_number.strftime("%M:%S") + "." + cur_number.strftime("%f")[:3])
    
    if len(x)>20 :
        x.pop(0)
        y.pop(0)
        #xcaption.pop(0)
    
    ax.clear()
    ax.plot(x, y)
    #plt.xticks(x, y, rotation ='vertical')
    plt.grid(color='gray', linestyle='-', linewidth=1.5) # включить отрисовку сетки c определенными параметрами

ani = an.FuncAnimation(fig=fig, func=UpdateGraph, frames=1, interval=1000)
plt.show()


```

```python
x = [1, 2, 3, 4, 5, 6]
y1 = [7, 13, 16, 18, 25, 19]
y2 = [29, 25, 20, 25, 20, 18]
y3 = [29, 25, 20, 25, 20, 18]
y4 = [29, 25, 20, 25, 20, 18]

#define grid of plots
fig2, ax2 = plt.subplots(nrows= 2 , ncols= 2 )

#add title
fig2.suptitle('Plots Stacked Vertically')
i=0;
 
def UpdateGraph2(frame):
    global i
    i=i+1
    x.append(i)
    y1.append(random.randint(5,10))
    y2.append(random.randint(0,5))
    y3.append(random.randint(-5,5))
    y4.append(random.randint(10,30))
    
    if len(x)>20 :
        x.pop(0)
        y1.pop(0)
        y2.pop(0)
        y3.pop(0)
        y4.pop(0)
    
    ax2[0,0].clear()
    ax2[0,1].clear()
    ax2[1,0].clear()
    ax2[1,1].clear()
    
    ax2[0,0].plot(x, y1)
    ax2[0,1].plot(x, y2)
    ax2[1,0].plot(x, y3)
    ax2[1,1].plot(x, y4)

    #plt.grid(color='gray', linestyle='-', linewidth=1.5) # включить отрисовку сетки c определенными параметрами

ani2 = an.FuncAnimation(fig=fig2, func=UpdateGraph2, frames=1, interval=1000)
plt.show()

#add data to plots
#ax2[0, 0].plot(x, y1)
#ax2[0, 1].plot(x, y2)
#ax2[1,0].plot(x, y3)
#ax2[1,1].plot(x, y4)
```
